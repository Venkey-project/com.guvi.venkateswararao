```python
from mysql import connector

# connecting to sql database
connection = connector.connect(
    host = "localhost",
    user = "root",
    password = "root",
    database = "PowerBi"
)
my_cursor = connection.cursor()
my_cursor
```

```python
query = "show tables"
my_cursor.execute(query)
for tb in my_cursor:
    print(tb)
```

```python
# sql query for sales data
query = "describe sales"
my_cursor.execute(query)
for db in my_cursor:
    print(db)
```

```python
query = "select COUNT(Quantity) from sales "
my_cursor.execute(query)
for db in my_cursor:
    print(db)
```

```python
query = "select DISTINCT(ProductKey) from sales GROUP BY Currency_Code,ProductKey"
my_cursor.execute(query)
for db in my_cursor:
    print(db)
```

```python
query = "select DISTINCT(CustomerKey) from sales GROUP BY ProductKey,CustomerKey"
my_cursor.execute(query)
for db in my_cursor:
    print(db)
```
